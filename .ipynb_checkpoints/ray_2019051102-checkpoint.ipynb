```python
import os
import re
from joblib import Parallel, delayed
from math import sqrt
```

```python
all_mail = {}
all_mail_code = set()
with open("D:\\TTS1.000", 'r') as mfile:
    for line in mfile:
        (mail_status, mail_code, mail_date, mail_time,post_code, others) = line.strip().split(",")
        
        if not mail_code in all_mail_code:
            all_mail_code.add(mail_code)
            all_mail[mail_code] = []
          
        all_mail[mail_code].append((mail_status, post_code, mail_date, mail_time))

```
